```python
def cyclin_sites(idr_dict):
    #finds the total number of Cdk Cyclin sites in an IDR and returns as a dictionary
    #motif is S/T, P, X, (R, K)
    cyclin_dict = {}
    for key in idr_dict:
        seq = idr_dict[key]
        length = len(seq)
        count = 0
        while seq:
            if (seq[0] == 'S' or seq[0] == 'T') and len(seq) >= 4:
                if seq[1] == 'P':
                    if seq[3] == 'R' or seq[3] == 'K':
                        count += 1
                        seq = seq[4:]
                    else:
                        seq = seq[1:] 
                else:
                    seq = seq[1:]
            else:
                seq = seq[1:]
        cyclin_dict[key] = [count, length]
    return cyclin_dict

def nonoptimal_cyclin_sites(idr_dict):
    #motif is just S/T, P
    nonoptimal_cyclin_dict = {}
    for key in idr_dict:
        seq = idr_dict[key]
        length = len(seq)
        count = 0
        while seq:
            if (seq[0] == 'S' or seq[0] == 'T') and len(seq) >= 2:
                if seq[1] == 'P':
                    count += 1
                    seq = seq[2:]
                else:
                    seq = seq[1:]
            else:
                seq = seq[1:]
        nonoptimal_cyclin_dict[key] = [count, length]
    return nonoptimal_cyclin_dict
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
import seaborn as sns
import pandas as pd
from sklearn.linear_model import LinearRegression

def dictionary_lister(dict):
    #takes in dictionaries with the following entry form:
    #name: [item 1, item 2]
    items1 = []
    items2 = []
    for key in dict:
        items1.append(dict[key][0])
        items2.append(dict[key][1])
    return items1, items2

def dataframe_generator(dict):
    #takes in the lists generated by dictionary_lister and puts the data in a dataframe
    count, length = dictionary_lister(dict)
    return pd.DataFrame({'Count': count, 'Length': length})

#This will be used for statistical analysis code (Ex. linear regression)
#https://medium.com/@mjfstanford/simple-linear-regression-in-python-905b759ef0e6
#slope and y-intercept for linear regression
#x and y must be in the numpy array form: np.array([list])

def slope(x, y):
    denominator = x.dot(x) - x.mean() * x.sum()
    return (x.dot(y) - y.mean() * x.sum()) / denominator

def y_intercept(x, y):
    denominator = x.dot(x) - x.mean() * x.sum()
    return (y.mean() * x.dot(x) - x.mean() * x.dot(y)) / denominator

#y_pred = slope(np.array(x), np.array(y)) * np.array(x) + y_intercept(np.array(x), np.array(y)) to create line
#plt.plot(x, y_pred) to plot line
#print('slope', slope(np.array(x), np.array(y)), 'y-intercept', y_intercept(np.array(x), np.array(y))) displays info

#Pearson correlation: print('r', np.corrcoef(np.array(x), np.array(y)))
```

```python
y, x = dictionary_lister(optimal_cyclin_sites)
y_pred = slope(np.array(x), np.array(y)) * np.array(x) + y_intercept(np.array(x), np.array(y))
plt.scatter(x, y)
plt.plot(x, y_pred)
plt.title('Optimal Sites to Length')
plt.xlabel('Length')
plt.ylabel('Number of Optimal Sites')
plt.show()
print('slope', slope(np.array(x), np.array(y)), 'y-intercept', y_intercept(np.array(x), np.array(y)))
print('r', np.corrcoef(np.array(x), np.array(y)))
```

```python
def optimal_cdk_location(idr_dict):
    sites_dict = {}
    for key in idr_dict:
        seq = idr_dict[key]
        length = len(seq)
        site = 1
        while seq:
            if (seq[0] == 'S' or seq[0] == 'T') and len(seq) >= 4:
                if seq[1] == 'P':
                    if seq[3] == 'R' or seq[3] == 'K':
                        if key not in sites_dict:
                            sites_dict[key] = [site]
                        else:
                            sites_dict[key].append(site)
                        site += 4
                        seq = seq[4:]
                    else:
                        site += 1
                        seq = seq[1:] 
                else:
                    site += 1
                    seq = seq[1:]
            else:
                site += 1
                seq = seq[1:]
    return sites_dict

def suboptimal_cdk_location(idr_dict):
    sites_dict = {}
    for key in idr_dict:
        seq = idr_dict[key]
        length = len(seq)
        site = 1
        while seq:
            if (seq[0] == 'S' or seq[0] == 'T') and len(seq) >= 2:
                if seq[1] == 'P':
                    if key not in sites_dict:
                        sites_dict[key] = [site]
                    else:
                        sites_dict[key].append(site)
                    site += 2
                    seq = seq[2:]
                else:
                    site += 1
                    seq = seq[1:]
            else:
                site += 1
                seq = seq[1:]
    return sites_dict
```
