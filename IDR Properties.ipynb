```python
from Bio.SeqUtils.ProtParam import ProteinAnalysis
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
import pandas as pd

def pI_dict_analysis(idr_dict):
    #takes in a dictionary with protein sequences as values and makes a new dictionary with pI
    pI_dict = {}
    for key in idr_dict:
        pI = ProteinAnalysis(idr_dict[key]).isoelectric_point()
        pI_dict[key] = pI
    return pI_dict

def dictionary_to_list(dict):
    data_points = []
    for key in dict:
        data_points.append(dict[key])
    return data_points

idr_pI_dict = pI_dict_analysis(idr_dict)
idr_pI_list = dictionary_to_list(idr_pI_dict)
print(idr_pI_dict)

data_to_plot = [idr_pI_list]

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

bp = ax.boxplot(data_to_plot, labels=['pI'])

plt.ylabel('pI')

#plt.savefig('pI Boxplot.png', bbox_inches='tight')
```

```python
def dict_length(idr_dict):
    length_dict = {}
    for key in idr_dict:
        length = len(idr_dict[key])
        length_dict[key] = length
    return length_dict

def dictionary_lister(dict):
    keys = []
    values = []
    for key in dict:
        keys.append(key)
        values.append(dict[key])
    return keys, values

x, y = dictionary_lister(dict_length(idr_dict))
plt.bar(x, y)
plt.xticks(rotation = 90)
plt.xlabel('Organism Name')
plt.ylabel('Number of Amino Acids')
plt.suptitle('Amino Acid Lengths of Various IDRs')

fig = plt.gcf()
fig.set_size_inches(18, 10.5)

#plt.savefig('IDR Lengths.png', bbox_inches='tight')
```

```python
def dict_to_dataframe(dict):
    keys, values = dictionary_lister(dict)
    return pd.DataFrame({'Organisms': keys, 'Length': values})

length_frame = dict_to_dataframe(dict_length(idr_dict))
sb.histplot(data=length_frame, x='Length', stat='density', fill=False, color='black')
```

```python
def fraction_composition(sequence, aa_list):
    #takes in sequence and finds percent composition of the given amino acids
    total = 0
    length = len(sequence)
    while sequence:
        if sequence[0] in aa_list:
            total += 1
            sequence = sequence[1:]
        else:
            sequence = sequence[1:]
    return total / length

def dict_composition(idr_dict, aa_list):
    output = []
    for key in idr_dict:
        output.append(fraction_composition(idr_dict[key], aa_list))
    return output

bulky_hydrophobics = dict_composition(idr_dict, ['L', 'I', 'V', 'M'])
aromatics = dict_composition(idr_dict, ['F', 'Y', 'W'])
charged = dict_composition(idr_dict, ['K', 'R', 'D', 'E', 'H'])
proline = dict_composition(idr_dict, ['P'])
hydrophilic = dict_composition(idr_dict, ['S', 'T', 'Q', 'N'])
small_hydrophobic = dict_composition(idr_dict, ['A', 'G', 'C'])

data_to_plot = [bulky_hydrophobics, aromatics, charged, proline, hydrophilic, small_hydrophobic]

fig = plt.figure()

ax = fig.add_axes([0,0,2,2])

bp = ax.boxplot(data_to_plot, vert=True,
        labels=['Bulky Hydrophobics', 'Aromatics', 'Charged', 'Proline', 'Hydrophilic', 'Small Hydrophobics'])

plt.xlabel('Percentage')
plt.title('Amino Acid Percentages of Various Organisms')

plt.show()
```

```python
def sb_dict_composition(idr_dict, aa_list, grouping_name):
    output = []
    for key in idr_dict:
        output.append([grouping_name, fraction_composition(idr_dict[key], aa_list), key])
    return output

sb_bulky_hydrophobics = sb_dict_composition(idr_dict, ['L', 'I', 'V', 'M'], 'Bulky Hydrophobics')
sb_aromatics = sb_dict_composition(idr_dict, ['F', 'Y', 'W'], 'Aromatics')
sb_positive = sb_dict_composition(idr_dict, ['K', 'R', 'H'], 'Positives')
sb_negative = sb_dict_composition(idr_dict, ['D', 'E'], 'Negatives')
sb_proline = sb_dict_composition(idr_dict, ['P'], 'Proline')
sb_hydrophilic = sb_dict_composition(idr_dict, ['S', 'T', 'Q', 'N'], 'Hydrophilics')
sb_small_hydrophobic = sb_dict_composition(idr_dict, ['A', 'G', 'C'], 'Small Hydrophobics')

data = sb_bulky_hydrophobics + sb_aromatics + sb_positive + sb_negative + sb_proline + sb_hydrophilic + sb_small_hydrophobic
labels = ['Amino Acid Type', 'Percentage', 'Organism']
aa_type_ratios_table = pd.DataFrame(data, columns=labels, dtype=float)
print(aa_type_ratios_table)
```
