```python
from localcider.sequenceParameters import SequenceParameters as sp

def kappa_dict(idr_dict):
    #prints out dict of idr with kappa
    k_dict = {}
    for idr in idr_dict:
        sequence_parameters = sp(idr_dict[idr])
        kappa = sequence_parameters.get_kappa()
        k_dict[idr] = kappa
    return k_dict
```

```python
kappa_dictionary = kappa_dict(idr_dict)
print(kappa_dictionary)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
def kappa_lister(kappa_dict):
    output = []
    for key in kappa_dict:
        output.append([key, kappa_dict[key], 'Cdt1'])
    return output

sb_kappa = kappa_lister(kappa_dictionary)

data = sb_kappa
labels = ['Organism', 'Kappa', 'Protein']
kappa_table = pd.DataFrame(data, columns=labels, dtype=float)

fig = plt.gcf()
fig.set_size_inches(12, 8)
sb.violinplot(x = 'Protein', y = 'Kappa', data = kappa_table, color='white')
#plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
#plt.title('Kappa Values of Multiple Organisms')

plt.savefig('Kappa Values of Idrs.png')
```
