```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
import pandas as pd

def residue_distance(protein_dictionary, residue_list):
    #finds the average distance between residues specified in the residue_list in amino acids
    average = {}
    for protein in protein_dictionary:
        distance = 0
        sequence = protein_dictionary[protein]
        values = []
        end = False
        if len(sequence) > 100:
            if sequence[-1] in residue_list:
                end = True
            while sequence:
                if sequence[0] in residue_list:
                    values.append(distance)
                    distance = 0
                    sequence = sequence[1:]
                else:
                    distance += 1
                    sequence = sequence[1:]
            if end:
                average[protein] = values[1:]
            elif len(values) <= 2:
                average[protein] = []
            else:
                average[protein] = values[1:len(values)-1]
    return average

def dictionary_organizer(dict, protein_dict, name):
    #this takes the list of values and organizes them into a new dictionary
    organized_dict = {}
    for key in dict:
        zero_count = 0
        nonzero_count = 0
        total = 0
        values = dict[key]
        while values:
            if values[0] == 0:
                zero_count += 1
                values = values[1:]
            else:
                nonzero_count += 1
                total += values[0]
                values = values[1:]
        if nonzero_count > 0:
            organized_dict[key] = [name, key, zero_count / (len(protein_dict[key]) / 100), total / nonzero_count]
        else:
            organized_dict[key] = [name, key, zero_count / (len(protein_dict[key]) / 100), total]
    return organized_dict

def dict_to_data_frame(organized_dict):
    #Only works with the dictionary_organizer function
    data_list = []
    for key in organized_dict:
        data_list.append(organized_dict[key])
    labels = ['Name', 'IDR', 'Zero', 'Non-Zero']
    table = pd.DataFrame(data_list, columns=labels, dtype=float)
    return table
```

```python
idr_large_hydrophobic_average = residue_distance(idr_dict, ['L', 'I', 'V', 'F', 'Y', 'W', 'M'])
large_hydrophobic = dictionary_organizer(idr_large_hydrophobic_average, idr_dict, 'Large Hydrophobic')

idr_small_hydrophobic_average = residue_distance(idr_dict, ['A', 'G', 'C'])
small_hydrophobic = dictionary_organizer(idr_small_hydrophobic_average, idr_dict, 'Small Hydrophobic')

idr_positive_average = residue_distance(idr_dict, ['R', 'K', 'H'])
positive = dictionary_organizer(idr_positive_average, idr_dict, 'Positive')

idr_negative_average = residue_distance(idr_dict, ['D', 'E'])
negative = dictionary_organizer(idr_negative_average, idr_dict, 'Negative')

idr_hydrophilic_average = residue_distance(idr_dict, ['S', 'T', 'N', 'Q'])
hydrophilic = dictionary_organizer(idr_hydrophilic_average, idr_dict, 'Hydrophilic')

idr_proline_average = residue_distance(idr_dict, ['P'])
proline = dictionary_organizer(idr_proline_average, idr_dict, 'Proline')

large_hydrophobic_table = dict_to_data_frame(large_hydrophobic)
small_hydrophobic_table = dict_to_data_frame(small_hydrophobic)
positive_table = dict_to_data_frame(positive)
negative_table = dict_to_data_frame(negative)
hydrophilic_table = dict_to_data_frame(hydrophilic)
proline_table = dict_to_data_frame(proline)

tables = [large_hydrophobic_table, small_hydrophobic_table, positive_table, negative_table, hydrophilic_table, proline_table]
result = pd.concat(tables)

fig1 = plt.gcf()
fig1.set_size_inches(12, 8)
sb.violinplot(x = 'Name', y = 'Zero', data = result, color='white')
plt.xlabel('Amino Acid Type')
plt.ylabel('Frequency')
#plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
#plt.title('Zero')
plt.savefig('Number of Adjacent Amino Acids in IDRs.png')
plt.show()

fig2 = plt.gcf()
fig2.set_size_inches(12, 8)
sb.violinplot(x = 'Name', y = 'Non-Zero', data = result, color='white')
plt.xlabel('Amino Acid Type')
plt.ylabel('Average Distance')
#plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
#plt.title('Non-Zero')
plt.savefig('Distance between Nonadjacent Amino Acids in IDRs.png')
plt.show()

#make this format consistent with the drosophila distance graphs
```
