```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
import pandas as pd
import random

def random_dict_selector(prot_dict, num, length):
    #selects num random proteins from a dict and puts them into a new dict
    #must be a certain length
    output = {}
    while num:
        protein = list(prot_dict)[random.randint(0, len(prot_dict)-1)]
        if protein not in output and len(prot_dict[protein]) >= length:
            output[protein] = prot_dict[protein]
            num -= 1
    return output

def random_nested_dict_selector(nested_prot_dict, num, length1, length2):
    output = {}
    while num:
        protein = list(nested_prot_dict)[random.randint(0, len(nested_prot_dict)-1)]
        if protein not in output:
            nest = {}
            for pos in nested_prot_dict[protein]:
                if len(nested_prot_dict[protein][pos]) >= length1 and len(nested_prot_dict[protein][pos]) <= length2:
                    nest[pos] = nested_prot_dict[protein][pos]
                    num -= 1
            if nest:
                output[protein] = nest
    return output

def fraction_composition(sequence, aa_list):
    #takes in sequence and finds percent composition of the given amino acids
    total = 0
    length = len(sequence)
    while sequence:
        if sequence[0] in aa_list:
            total += 1
            sequence = sequence[1:]
        else:
            sequence = sequence[1:]
    return total / length

def dict_composition(prot_dict, aa_list, name, region):
    output = []
    for key in prot_dict:
        output.append([name, region, fraction_composition(prot_dict[key], aa_list)])
    return output

def nested_dict_composition(nested_prot_dict, aa_list, name, region):
    output = []
    for key in nested_prot_dict:
        for pos in nested_prot_dict[key]:
            output.append([name, region, fraction_composition(nested_prot_dict[key][pos], aa_list)])
    return output
```
